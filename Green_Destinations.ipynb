# 1. Import And Inspect Data by using Python

```python
from google.colab import files
uploaded = files.upload()
import pandas as pd
import io
df = pd.read_csv(io.BytesIO(uploaded[list(uploaded.keys())[0]]))
df.head()
df.info()
df.describe()
```

# 2. Calculating the Attributes Rates


```python
attrition_rate = (df['Attrition'].value_counts()['Yes'] / len(df)) * 100
print(f"Attrition Rate: {attrition_rate:.2f}%")
```

# 3. Exporing Relationships

A. Age v/s Attrition

```python
df.groupby('Attrition')['Age'].mean()
```

Plot:

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(x='Attrition', y='Age', data=df)
plt.title('Age vs Attrition')
plt.show()
```

B. Years at Company v/s Attrition

```python
sns.boxplot(x='Attrition', y='YearsAtCompany', data=df)
plt.title('Years at Company vs Attrition')
plt.show()
```

C. Income v/s Attrition

```python
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title('Income vs Attrition')
plt.show()
```

# 4. Correlation or Predictive Insights

```python
corr = df[['Age', 'YearsAtCompany', 'MonthlyIncome']].corr()
print(corr)
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

# 5. Insights



*   Overall attrition rate
*   Average age, income, and tenure of employees who left vs stayed
*   Insights (Younger employee with fewer years at company tend to leave more often)
*   Recommendations for HR to reduce attrition
