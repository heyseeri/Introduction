```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt


from google.colab import files
uploaded = files.upload()

df = pd.read_csv("greendestination (1).csv")



# Attrition rate
attrition_rate = df['Attrition'].value_counts(normalize=True) * 100
print("Attrition Rate (%):")
print(attrition_rate)

# Compare factors (Age, Years at Company, Monthly Income)
attrition_factors = df.groupby('Attrition')[['Age','YearsAtCompany','MonthlyIncome']].mean()
print("\nAverage factors by Attrition status:")
print(attrition_factors)



plt.style.use("ggplot")

# 1. Attrition Rate Pie Chart
attrition_counts = df['Attrition'].value_counts()
plt.figure(figsize=(5,5))
plt.pie(attrition_counts, labels=attrition_counts.index, autopct='%1.1f%%', startangle=90, colors=['#4CAF50','#F44336'])
plt.title("Attrition Rate")
plt.show()

# 2. Age vs Attrition
plt.figure(figsize=(7,5))
df.boxplot(column='Age', by='Attrition', grid=False, patch_artist=True)
plt.title("Age vs Attrition")
plt.suptitle("")
plt.ylabel("Age")
plt.show()

# 3. Years at Company vs Attrition
plt.figure(figsize=(7,5))
df.boxplot(column='YearsAtCompany', by='Attrition', grid=False, patch_artist=True)
plt.title("Years at Company vs Attrition")
plt.suptitle("")
plt.ylabel("Years at Company")
plt.show()

# 4. Monthly Income vs Attrition
plt.figure(figsize=(7,5))
df.boxplot(column='MonthlyIncome', by='Attrition', grid=False, patch_artist=True)
plt.title("Monthly Income vs Attrition")
plt.suptitle("")
plt.ylabel("Monthly Income")
plt.show()
```
